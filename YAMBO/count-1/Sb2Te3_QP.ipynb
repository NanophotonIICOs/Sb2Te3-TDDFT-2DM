```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import TABLEAU_COLORS, same_color
import pickle
import sys
import glob 
import os
import re

def find_data(datadir,wdet):
    data = []    
    for (root, dirs, files) in os.walk(datadir):
        for file in files:
            if wdet in file:
                file_path = os.path.join(root, file)
                data.append(file_path)
    return data

def extract_number(file):
    m = re.search(r"ox_([+-]?\d+\.\d+)", file)
    return float(m.group(1)) if m else 0.0

def extract_number2(file):
    m = re.search(r"o1-(\d+\.\d+)", file)
    return float(m.group(1)) if m else 0.0 
```

# G0W0 YAMBO

```python
ks = np.loadtxt("QP-BSE/g0w0-bands-out/o-g0w0_bands_3_run.bands_interpolated_01")
g0w0_4 = np.loadtxt("QP-BSE/g0w0-bands-out/o-g0w0_bands_4_run.bands_interpolated")
g0w0_5 = np.loadtxt("QP-BSE/g0w0-bands-out/o-g0w0_bands_5_run.bands_interpolated")
g0w0_6 = np.loadtxt("QP-BSE/g0w0-bands-out/o-g0w0_bands_6_run.bands_interpolated")
g0w0_7 = np.loadtxt("QP-BSE/g0w0-bands-out/o-g0w0_bands_7_run.bands_interpolated")

plt.figure(figsize=(5,7))
for i in range(1,15):
    plt.plot(ks[:,0],ks[:,i],"k",label=r"GGA" if i == 1 else "")
for i in range(1,15):
    plt.plot(g0w0_4[:,0],g0w0_4[:,i],"m",label=r"G0W0 calc-4" if i == 1 else "")
    plt.plot(g0w0_5[:,0],g0w0_5[:,i],"y",label=r"G0W0 calc-5" if i == 1 else "")
    plt.plot(g0w0_6[:,0],g0w0_6[:,i],"c",label=r"G0W0 calc-6" if i == 1 else "")
    plt.plot(g0w0_7[:,0],g0w0_7[:,i],"r",label=r"G0W0 calc-4" if i == 1 else "")

plt.ylim(-2,2)
plt.xlim(0,ks[:,0][-1])
plt.xticks([0,0.51,ks[:,0][-1]],["k","G","M"])
plt.axvline(0.51,c='k',ls='--',linewidth=0.5, alpha=0.5)
plt.axhline(0.0,c='k',ls='--',linewidth=0.5, alpha=0.5)

plt.legend(loc=3)
plt.show()
```
